```python
import pandas as pd
# reading the csv file
customer_data = pd.read_csv('BankChurners.csv') 

# Displaying basic information about the dataset
customer_data.info()
```

```python
# Checking for empty cells (NaN values)

# Checking for missing values in each column
print(customer_data.isnull().sum())

# Alternatively, checking if any missing values exist
print(customer_data.isnull().any().any())  # Returns True if any NaN exists
```

```python
# Dropping attributes that are not required
columns_to_drop = ['CLIENTNUM',
                   'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
                   'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2']
customer_data = customer_data.drop(columns=columns_to_drop)
```

```python
# Analysing Card_Category V/S Total_Trans_Amt for all customers

# Group by 'Card_Category' and aggregate only numeric columns
grouped_df = (
    customer_data.groupby('Card_Category')
    .mean(numeric_only=True)  # Ensures only numeric columns are used
    .reset_index()
)

# Selecting only the Card_Category column and Total_Trans_Amt column
selected_columns = ['Card_Category', 'Total_Trans_Amt']
result_df_original = grouped_df[selected_columns]

print(result_df_original)

```

```python
# Mapping Attrition_Flag to desired values
# 'Existing Customer' => 0, 'Attrited Customer' => 1
customer_data['Attrition_Flag'] = customer_data['Attrition_Flag'].map({
    'Existing Customer': 0,
    'Attrited Customer': 1
})
```

```python
attrition_df = customer_data[customer_data['Attrition_Flag'] == 1.0]
existing_df = customer_data[customer_data['Attrition_Flag'] == 0]
```

```python
# Analysing Card_Category V/S Total_Trans_Amt for existing customers

# Group by 'Card_Category' and aggregate only numeric columns
grouped_df = (
    existing_df.groupby('Card_Category')
    .mean(numeric_only=True)  # Ensures only numeric columns are used
    .reset_index()
)

# Selecting only the Card_Category column and Total_Trans_Amt column
selected_columns = ['Card_Category', 'Total_Trans_Amt']
result_df_existing = grouped_df[selected_columns]

print(result_df_existing)
```

```python
# Analysing Card_Category V/S Total_Trans_Amt for attrited customers

# Group by 'Card_Category' and aggregate only numeric columns
grouped_df = (
    attrition_df.groupby('Card_Category')
    .mean(numeric_only=True)  # Ensures only numeric columns are used
    .reset_index()
)

# Selecting only the Card_Category column and Total_Trans_Amt column
selected_columns = ['Card_Category', 'Total_Trans_Amt']
result_df_attrited = grouped_df[selected_columns]

print(result_df_attrited)
```
