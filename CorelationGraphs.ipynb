# Importing Modules

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Computing Correlation Matrix

```python
# Selecting only numeric columns
df = pd.read_csv('BankChurners.csv')
df = df.drop(columns=['CLIENTNUM','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'])
numeric_df = df.select_dtypes(include=[np.number])

# Computing the correlation matrix
correlation_matrix = numeric_df.corr()

# Displaying the correlation matrix
correlation_matrix
```

# Comparing Churned vs. Retained Customers with Correlation Graphs

```python
# Loading the dataset
data = pd.read_csv('BankChurners.csv')

# Key features for analysis (excluding CLIENTNUM and Naive Bayes probabilities)
key_features = [
    'Customer_Age', 'Dependent_count', 'Months_Inactive_12_mon', 
    'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal', 
    'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt', 
    'Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio'
]

# Splitting data into churned and retained
churned = data[data['Attrition_Flag'] == 'Attrited Customer']
retained = data[data['Attrition_Flag'] == 'Existing Customer']

# Calculating descriptive statistics
churned_stats = churned[key_features].mean()
retained_stats = retained[key_features].mean()

comparison = pd.DataFrame({'Churned': churned_stats, 'Retained': retained_stats})
print(comparison)

# Visualizing the differences through bar graph
comparison.plot(kind='bar', figsize=(14, 7))
plt.title('Feature Comparison Between Churned and Retained Customers')
plt.ylabel('Mean Value')
plt.show()

# Boxplots for highly correlated features
high_corr_features = ['Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Avg_Utilization_Ratio', 'Total_Trans_Amt']
for feature in high_corr_features:
    sns.boxplot(x='Attrition_Flag', y=feature, data=data)
    plt.title(f'Distribution of {feature} by Attrition_Flag')
    plt.show()
```

# Creating Correlation Heatmap for the dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the dataset from the uploaded file
file_path = 'BankChurners.csv'
data = pd.read_csv(file_path)

# Converting target column to binary for correlation analysis
data['Attrition_Flag'] = data['Attrition_Flag'].apply(lambda x: 1 if x == 'Attrited Customer' else 0)
data = data.drop(columns=['CLIENTNUM','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'])

# Selecting numerical features for correlation heatmap
numerical_features = data.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = data[numerical_features].corr()

# Plotting the heatmap
plt.figure(figsize=(12, 10))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
