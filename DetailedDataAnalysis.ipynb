```python
import pandas as pd

# reading a csv file
df = pd.read_csv('BankChurners.csv') 
# Dropping unnecessary columns
df = df.drop(columns=['CLIENTNUM','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'])
```

```python
# Displaying the count of each category for all categorical columns
for i in df.select_dtypes(include = 'object').columns:
    print(df.groupby(by = [i])[i].count())
    print("\n")       
```

```python
# Creating 4 seperate columns from the card_category column
df = pd.get_dummies(df, columns=['Card_Category'], prefix='', prefix_sep='')

# Converting the newly created columns to integers
for col in ['Blue','Silver', 'Gold', 'Platinum']:
    if col in df.columns:
        df[col] = df[col].astype(int)

# Rearranging the columns for clarity
columns_order = [col for col in ['Blue',  'Silver','Gold', 'Platinum'] if col in df.columns] + \
                [col for col in df.columns if col not in ['Blue','Silver', 'Gold', 'Platinum']]
df = df[columns_order]

print(df.info())
```

```python
df.describe()
```

```python
# Group by 'Card_Category' and aggregate only numeric columns
grouped_df = (
    df.groupby('Income_Category')
    .mean(numeric_only=True)  # Ensures only numeric columns are used
    .reset_index()
)

# Select the desired columns using a list
selected_columns = ['Income_Category', 'Total_Trans_Amt']
result_df = grouped_df[selected_columns]

result_df

```

```python
grouped_df1 = (
    df.groupby('Income_Category')
    .mean(numeric_only=True)  # Ensures only numeric columns are used
    .reset_index()
)

selected_columns = ['Income_Category', 'Total_Trans_Amt','Blue','Silver','Gold','Platinum']
result_df = grouped_df1[selected_columns]

result_df
```

```python
atrition_mapping = {
    "Attrited Customer":0,
    "Existing Customer":1
}

df['Attrition_Flag'] = df['Attrition_Flag'].map(atrition_mapping)
df['Attrition_Flag'] = df['Attrition_Flag'].astype(float)
```

```python
# Filtering out records with utilization ratio less than or equal to 0.05 and calculate their Total_Trans_Amt and  Total_Trans_Ct -->

# Filter records where Avg_Utilization_Ratio is less than or equal to 0.05
filtered_df = df[df['Avg_Utilization_Ratio'] <= 0.05]

# # Calculate the total for Total_Trans_Amt and Total_Trans_Ct
total_trans_amt = filtered_df['Total_Trans_Amt'].sum()
total_trans_ct = filtered_df['Total_Trans_Ct'].sum()

print("Total Transaction Amount (Total_Trans_Amt):", total_trans_amt)
print("Total Transaction Count (Total_Trans_Ct):", total_trans_ct)
```

```python
# Filtering out records with utilization ratio greater than 0.05 and calculate their Total_Trans_Amt and  Total_Trans_Ct -->
# Filter records where Avg_Utilization_Ratio is greater than 0.05
majority_filtered_df = df[df['Avg_Utilization_Ratio'] > 0.05]

# # Calculate the total for Total_Trans_Amt and Total_Trans_Ct
total_trans_amt = majority_filtered_df['Total_Trans_Amt'].sum()
total_trans_ct = majority_filtered_df['Total_Trans_Ct'].sum()

print("Total Transaction Amount (Total_Trans_Amt):", total_trans_amt)
print("Total Transaction Count (Total_Trans_Ct):", total_trans_ct)
```
